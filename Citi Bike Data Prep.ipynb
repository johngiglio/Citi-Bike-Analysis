```python
import pandas as pd
```

Read and inspect data for 2018 citi bike trips and subway station locations

```python
jan18 = pd.read_csv('Data/201801-citibike-tripdata.csv')
feb18 = pd.read_csv('Data/201802-citibike-tripdata.csv')
mar18 = pd.read_csv('Data/201803-citibike-tripdata.csv')
apr18 = pd.read_csv('Data/201804-citibike-tripdata.csv')
may18 = pd.read_csv('Data/201805-citibike-tripdata.csv')
jun18 = pd.read_csv('Data/201806-citibike-tripdata.csv')
jul18 = pd.read_csv('Data/201807-citibike-tripdata.csv')
aug18 = pd.read_csv('Data/201808-citibike-tripdata.csv')
sep18 = pd.read_csv('Data/201809-citibike-tripdata.csv')
subways = pd.read_csv('Data/StationEntrances.csv')
```

Transform dataframes for mapping viz. Add station type columns, , drop duplicates, concatenate

```python
#Concatenate all 2018 bike station dfs into a single df
bike18 = pd.concat([jan18,feb18,mar18,apr18,may18,jun18,jul18,aug18,sep18], ignore_index=True)
```

```python
bike18.info()
```

```python
#export to csv
bike18.to_csv('Data/2018_citi_bike_trips.csv')
```

```python
#create df for just start station locations, rename columns
bike_start_stations_18 = bike18[['start station name', 'start station latitude',
                                 'start station longitude','tripduration', 'bikeid']]
bike_start_stations_18 = bike_start_stations_18.rename(columns = {'start station name':'Station Name',
                                                                  'start station latitude':'Station Latitude',
                                                                  'start station longitude':'Station Longitude',
                                                                 'tripduration':'Trip Duration'})
```

```python
bike_start_stations_18.head()
```

```python
#create df for just end station locations, rename columns
bike_end_stations_18 = bike18[['end station name', 'end station latitude',
                                 'end station longitude','tripduration', 'bikeid']]
bike_end_stations_18 = bike_end_stations_18.rename(columns = {'end station name':'Station Name',
                                                                  'end station latitude':'Station Latitude',
                                                                  'end station longitude':'Station Longitude',
                                                                 'tripduration':'Trip Duration',
                                                             'bikeid':'# of Trips'})
```

Group start and end station data frames by station to get total trip duration and trip count at each station

```python
bike_start_station_totals = pd.DataFrame(bike_start_stations_18.groupby('Station Name').agg({'Trip Duration':'sum','bikeid':'count','Station Latitude':'mean','Station Longitude':'mean'}))
bike_start_station_totals = bike_start_station_totals.rename(columns={'bikeid':'# of Trips'})
```

```python
bike_end_station_totals = pd.DataFrame(bike_end_stations_18.groupby('Station Name').agg({'Trip Duration':'sum','# of Trips':'count','Station Latitude':'mean','Station Longitude':'mean'}))
bike_end_station_totals.head()
```

```python
bike_start_station_totals = bike_start_station_totals.reset_index()
bike_end_station_totals = bike_end_station_totals.reset_index()
```

```python
bike_start_station_totals['Trip Duration'].describe()['75%']
```

```python
subway_marker_size1 = round(bike_start_station_totals['Trip Duration'].describe()['75%']*1.1)
```

```python
bike_start_station_totals['# of Trips'].describe()['75%']
```

```python
subway_marker_size2 = round(bike_start_station_totals['# of Trips'].describe()['75%']*1.1)
```

```python
subways = subways[['Station_Name','Station_Latitude','Station_Longitude']]
subways = subways.drop_duplicates()
subways['Station Type'] = 'SUBWAY'
```

```python
subways['Trip Duration'] = subway_marker_size1
subways['# of Trips'] = subway_marker_size2
```

```python
subways = subways.rename(columns ={'Station_Name':'Station Name',
                                   'Station_Latitude':'Station Latitude',
                                   'Station_Longitude':'Station Longitude'} )
```

```python
subways.head()
```

```python
bike_start_station_totals['Station Type'] = 'BIKE START'
bike_end_station_totals['Station Type'] = 'BIKE END'
bike_start_station_totals.head()
```

```python
all_stations_2018 = pd.concat([bike_start_station_totals,bike_end_station_totals,subways], ignore_index=True)
```
